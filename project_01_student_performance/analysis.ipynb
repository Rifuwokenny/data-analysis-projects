This project explores student performance data using NumPy.

## Part 1: Data Analysis with NumPy

```python
import numpy as np
```

# Creating NumPy Arrays

```python
marks = np.array([65, 72, 80, 58, 90])
marks
```

# Basic statistics

```python
print("Mean:", np.mean(marks))
print("Median:", np.median(marks))
print("Standard Deviation:", np.std(marks))
print("Max:", np.max(marks))
print("Min:", np.min(marks))
```

# Vectorised Operations

```python
adjusted_marks = marks + 5
adjusted_marks
```

# Logical conditions on arrays

```python
pass_mask = marks >= 50
pass_mask
```

```python
passed_students = marks[pass_mask]
passed_students
```

## Part 2: Data Analysis with Pandas

```python
import pandas as pd
```

## Table of data

```python
data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Marks": [65, 72, 80, 58, 90]
}

df = pd.DataFrame(data)
df
```

## Basic Inspection

```python
df.head()
```

```python
df.describe()
```

## Filtering data

```python
df[df["Marks"] >= 50]
```

```python
df[df["Marks"] < 50]
```

## Adding a new column

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 50 else "Fail")
df
```

## Mini Task Practice

```python
data = {
    "Student": [ "Rifuwo", "Kenny", "Junior", "Lucas", "Michelle"],
    "Marks": [ 69, 41, 85, 72, 33]
}

df = pd.DataFrame(data)
df
```

```python
mean_marks = df["Marks"].mean()
max_marks = df["Marks"].max()

mean_marks, max_marks
```

```python
def assign_grade(mark):
    if mark >= 75:
        return "A"
    elif mark >= 65:
        return "B"
    elif mark >= 50:
        return "C"
    else:
        return "Fail"
```

```python
df["Grade"] = df["Marks"].apply(assign_grade)
df
```
